```python
all_edges = [("A", "B"), ("A" "C"), ("B", "D"), ("C" "D"), ("C", "E"), ("D", "E")]
```

```python
adj_list = {
"A": [],
"B": [],
"C": [],
"D": [],
"E": []
}
```

```python
class Graph:
    def __init__(self, Nodes, is_directed=False):
        self.nodes = Nodes
        self.adj_list = {}
        self.is_directed = is_directed

        for node in self.nodes:
            self.adj_list[node] = []

    def add_edge(self, u, v):
        # we are creating a edge between 2 vertices
        # first we create edge between u and v
        self.adj_list[u].append(v)

        #if graph is not directed (it means relationship goes both ways and you can traverse A - B and B - A aswell)
        if not self.is_directed:
            # we create edge between v and u
            self.adj_list[v].append(u)

        

    def print_adj_list(self):
        for node in self.nodes:
            print(node, "->",  self.adj_list[node])

    #total number of edges and node to a given node
    def degree(self, node):
        deg = len(self.adj_list[node])
        return deg
        
        
        
    
        
    
```

```python
nodes = ["A", "B", "C", "D", "E"]
graph = Graph(nodes, is_directed=True )

#graph.print_adj_list()
for u, v in all_edges:
    graph.add_edge(u, v)
graph.print_adj_list()

#using degree function
print(graph.degree("A"))

    
    
```
